### Guardando os textos dos capítulos 17 e 8 com o BeautifulSoup (como na Atividade 02)

```
# Importa bibliotecas
import requests
from bs4 import BeautifulSoup
import re
```

```
# Busca a página HTML dos capítulos analisados (nesse caso, dos Capítulos 17 e 8)
resp_17 = requests.get('https://brasileiraspln.com/livro-pln/1a-edicao/parte8/cap17/cap17.html')
resp_08 = requests.get('https://brasileiraspln.com/livro-pln/1a-edicao/parte5/cap8/cap8.html')

# Utiliza o BS para obter um objeto que "entende" as tags HTML do conteúdo da requisição feita
soup_17 = BeautifulSoup(resp_17.content, 'html.parser')
soup_08 = BeautifulSoup(resp_08.content, 'html.parser')

print(soup_17)
print(soup_08)
```

```
# Procura pela tag main que contém a parte da página que queremos analisar
corpo_17 = soup_17.find_all('main', {'class': 'content'})
corpo_08 = soup_08.find_all('main', {'class': 'content'})

print(corpo_17)
print(corpo_08)
```

```
#Transforma o conteúdo textual da página em uma única string
texto_17 = ''
for x in corpo_17:
    texto_17 += x.getText()

texto_08 = ''
for x in corpo_08:
    texto_08 += x.getText()

print(texto_17)
print(texto_08)
```

Neste ponto, as variáveis texto_17 e texto_08 armazenam os textos das páginas referentes ao capítulos 17 e 8, respectivamente.

OBSERVAÇÃO: Como os textos dos capitulos são muito extensos, ao usar o chat completions, a API nos informa que o texto dos capítulos excede o número de tokens. Para contornar isso, dividirei os textos dos capítulos em várias partes e processarei separadamente.

A API diz que a soma do prompt e da saída pode ter no máximo 4097 tokens, e aponta que o texto do capítulo 17 tem aproximadamente 21500 tokens e o texto do capítulo 8 tem aproximadamente 6500 tokens. Assim, deixarei as entradas com cerca de 3000 tokens, para sobrar para a saída, caso ela também seja um texto extenso.

O texto do Capítulo 8 foi dividido em duas partes de pouco mais de 3000 tokens que, juntas, somam ele inteiro. Já o texto do capítulo 17, para não ficar uma tarefa repetitiva e que creio não ser o foco da atividade, foi dividido em duas partes, contendo cada uma cerca de 3000 tokens, mas está incompleto na análise.

Deixei dessa forma por achar que o foco não era quantidade, mas a ideia empregada em cada tópico!

## Primeira Técnica: Correção Gramatical

```
#@title Correção Gramatical usando o endpoint *Chat Completions* (Capítulo 17 - Parte 1)

import requests

endpoint = "https://api.openai.com/v1/chat/completions"

mensagem_sistema = 'Você é o editor de um livro a ser lançado em breve.'
mensagem_usuario = "Considere o seguinte texto: " + texto_17_parte_1
mensagem_assistente = "Primeiro, liste todos os erros gramaticais encontrados, apontando caso não tenha encontrado nenhum. Depois, indique outras possíveis correções, citando exemplos."

parametros = {
   "model": "gpt-3.5-turbo-0613",
   "messages": [
      {"role": "system", "content": mensagem_sistema},
      {"role": "user", "content": mensagem_usuario},
      {"role": "assistant", "content": mensagem_assistente}]
}
headers = {
   "Content-Type": "application/json",
   "Authorization": f"Bearer {openai.api_key}"
}

resposta = requests.post(endpoint, json=parametros, headers=headers)
resposta.json()
print(resposta.json()["choices"][0]["message"]["content"])
```
